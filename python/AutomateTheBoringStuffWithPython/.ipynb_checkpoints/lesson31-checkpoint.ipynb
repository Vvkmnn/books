### Reading and Writing Plaintext Files


__Plaintext files__ contain basic text characters, and no other information. Scripts are typically text.

__Binary files__ are any other kind of file, and are usually designed to be open by specific programs.

```python
import os

os.listdir(os.path.abspath('files'))
```

The `open()` function opens a given filename as an object in '_read mode_'.

It can than have methods called on it, like `.read()` and `.readline()`. 

_Read mode_ means that files can only be read in this stage; not re-written/appended.

Files must be closed with `.close()` to save changes. 

```python
dictionaryFile = open(os.path.abspath('files/dictionary.txt')) # Open the file

print(dictionaryFile.read()) # Read the file
print(dictionaryFile.readline()) # Read a line in the file until newline or EOF

content = dictionaryFile.read() # Store the content in a variable
```

To open a file in _write_ mode, pass the `w` parameter to the `open()` function.

This rewrites the entire file, and can also create files. 

```python
helloFile = open(os.path.abspath('files/helloworld.txt'), 'w') # Create the file in writeable mode.

helloFile.write('Hello World!') # Writes to file, and returns bytes used exactly as is (no newline)

helloFile.close() # Closes/saves files

helloFile = open(os.path.abspath('files/helloworld.txt'), 'r') # Open the file in read mode

print(helloFile.read()) # Print file contents

helloFile = open(os.path.abspath('files/helloworld.txt'), 'w') # Open file for rewriting

helloFile.write('Hello Another World!') # Writes to file, and returns bytes used exactly as is (no newline)

helloFile.close() # Closes/saves files

helloFile = open(os.path.abspath('files/helloworld.txt'), 'r') # Open the file in read mode

print(helloFile.read()) # Final file contents
```

Unless otherwise specified, all new files are created in the current working directory.

```python
os.getcwd()
```

To append a file instead of overwriting it, use _append_ mode by passing the `a` parameter to the `open()` function.

```python
baconFile = open(os.path.abspath('files/bacon.txt'), 'w') # Create file in write mode

baconFile.write('Bacon is not a vegetable.') # Write string to file

baconFile.close() # Save/close file

baconFile = open(os.path.abspath('files/bacon.txt'), 'r') # Reopen file in read mode

print(baconFile.read()) # Read the current file

baconFile = open(os.path.abspath('files/bacon.txt'), 'a') # Reopen file in append mode

baconFile.write('\nBacon is delicious.') # Write string to file

baconFile.close() # Save/close file

baconFile = open(os.path.abspath('files/bacon.txt'), 'r') # Reopen file in read mode

print(baconFile.read()) # Read the appended file
```

```python
import shelve # import the shelve module

shelfFile = shelve.open('mycatdata')

# Create a shelf file that works like a dictionary

shelfFile['cats'] = ['Zophie', 'Pooka', 'Simon', 'Fat-tail', 'Cleo']

shelfFile.close()
```

`shelve` files work like dictionaries, but are stored in binary files, which cannot be easily understood in text editors

```python
shelfFile = shelve.open('mycatdata') # Open shelve file

list(shelfFile.keys())

list(shelfFile.values())
```

## Recap
* `open()` will return a file has reading and writing-related methods.
* Pass `r` or nothing to open() to open the file in read, mode, `w` for write mode, and `a` for append mode.
* Opening a nonexistent filename in write or append mode will creat ethat file.
* Call `.read()` or `.write()` to read the contents of a file or write a string to a file.
* Call `.readlines()` to return a list of strings of that file's contents.
* Call `.close()` when you are done with a file. 
* The `shelve` module can store Python values in a binary file.
* `shelve.open()` returns a dictionary-like shelf value.
